```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from matplotlib import style
from scipy import stats
from scipy.stats import expon
from scipy.stats import norm

# In the following line we read data from the CSV file.
# 1-First we define number of columns in our CSV data sheet and give them descriptive names.
# 2-First row in our sheet is garbage so we need to drop using "skiprows" function.
# 3-"names" function assigns names to each column defined by the user. 

col = ['Time_100', 'Data_100', 'Time_1000', 'Data_1000', 'Time_10000', 'Data_10000']
data = pd.read_csv('lcg_ex3.csv', 0, delimiter = ',', skiprows = [0], names = col)


# After extracting data in columns we need to removes "NaN" values whihc Python automatically assigns to places 
# where there is no data. Since our coulumns are of different length so the first two contain 9900 NaN values 
# and subsequent couple contains 9000 NaN values, we only keep unique values here. 

X1 = data['Data_100'].dropna().unique()
X2 = data['Data_1000'].dropna().unique()
X3 = data['Data_10000'].dropna().unique()
Y1 = data['Time_100'].dropna().unique()
Y2 = data['Time_1000'].dropna().unique()
Y3 = data['Time_10000'].dropna().unique()


# Now we define our plotting area and figure sizes, since we need three histograms and three plots so we define them here. 


fig, axes = plt.subplots(3, 3, figsize = (20,15))

# To adjust horizontal distance between subplots
# matplotlib.pyplot.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

fig.subplots_adjust(hspace = 0.2 ) 



# Here we plot our histograms in the first row and three columns. 
# Each of them is plotted in different color and bin size is 
# kept 'auto' so that system automatically adjusts values. Note: ec= edge color

axes[0, 0].hist(X1, bins = 'auto', color = "red", ec="green"); axes[0,0].set_title("100 Simulation");axes[0,0].legend(["LCG RNG"])
axes[0, 1].hist(X2, bins = 'auto', color = "blue", ec="red"); axes[0,1].set_title("1K Simulation");axes[0,1].legend(["LCG RNG"])
axes[0, 2].hist(X3, bins = 'auto', color = "green", ec="blue"); axes[0,2].set_title("10K Simulation");axes[0,2].legend(["LCG RNG"])


# Here we plot Probabilty Density Functions in the second row and three columns for the given data. 

# The argument in the second and the third place are location and scale, also known as mean and standard deviation respectively.
# 100 Simulations
y1 = expon.pdf(X1, 0, 1.088)
axes[1, 0].plot(X1,y1,'r.', lw=5, alpha=0.6, color = "red"); axes[1,0].set_title("Exponential PDF_100");axes[1,0].legend(["LCG RNG"])
#1000 Simulations
y2 = expon.pdf(X2, 0, 0.967)
axes[1, 1].plot(X2,y2,'r.', lw=5, alpha=0.6, color = "blue"); axes[1,1].set_title("Exponential PDF_1K");axes[1,1].legend(["LCG RNG"])
#10000 Simulations
y3 = expon.pdf(X3, 0, 1.013)
axes[1, 2].plot(X3,y3,'r.', lw=5, alpha=0.6, color = "green" ); axes[1,2].set_title("Exponential PDF_10K");axes[1,2].legend(["LCG RNG"])


# Here we shall plot Cumulative Distribution Function(CDF).

# 100 Simulations
y1 = expon.cdf(X1, 0, 1.088)
axes[2, 0].plot(X1,y1,'r.', lw=5, alpha= 0.6, color = "red"); axes[2,0].set_title("Exponential CDF_100");axes[2,0].legend(["MT RNG"], loc='lower right')
#1000 Simulations
y2 = expon.cdf(X2, 0, 0.967)
axes[2, 1].plot(X2,y2,'r.', lw=5, alpha= 0.6, color = "blue"); axes[2,1].set_title("Exponential CDF_1K");axes[2,1].legend(["MT RNG"], loc='lower right')
#10000 Simulations
y3 = expon.cdf(X3, 0, 1.013)
axes[2, 2].plot(X3,y3,'r.', lw=5, alpha= 0.6, color = "green" ); axes[2,2].set_title("Exponential CDF_10K");axes[2,2].legend(["MT RNG"], loc='lower right')
```
