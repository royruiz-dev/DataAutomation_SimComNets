# The following code distributes the points uniformly inside a circle

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

R = 5
num_points = 10000

# Generate radius and theta from independent random number generators
radius = np.random.uniform(0, R, num_points)**0.5
theta = np.random.uniform(0, 2*np.pi, num_points)

# Map polar coordinates to cartesian
x = radius*np.cos(theta)
y = radius*np.sin(theta)

# Plot these points
plt.figure(figsize=(5,5))
plt.scatter(x, y, marker = '.', s = 1)
```
