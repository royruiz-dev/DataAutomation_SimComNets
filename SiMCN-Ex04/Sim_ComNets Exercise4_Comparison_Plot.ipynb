```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from matplotlib import style


col = ['Sim_T', 'Data']
data = pd.read_csv('csv_files/10ToMinus5BER.csv', delimiter = ',', skiprows = [0], names = col)
data1 = pd.read_csv('csv_files/10ToMinus7BER.csv', delimiter = ',', skiprows = [0], names = col)
data2 = pd.read_csv('csv_files/10ToMinus7_MHDR.csv', delimiter = ',', skiprows = [0], names = col)
data3 = pd.read_csv('csv_files/10ToMinus7_DD.csv', delimiter = ',', skiprows = [0], names = col)
data4 = pd.read_csv('csv_files/10ToMinus7_HD.csv', delimiter = ',', skiprows = [0], names = col)

X1 = data['Sim_T']
Y1 = data['Data']

X2 = data1['Sim_T']
Y2 = data1['Data']

X3 = data2['Sim_T']
Y3 = data2['Data']

X4 = data3['Sim_T']
Y4 = data3['Data']

X5 = data4['Sim_T']
Y5 = data4['Data']

plt.figure(figsize=(20,10))
plt.plot(X1,Y1,'r-', lw = 2, alpha = 0.8, color = "red"); 
plt.plot(X2,Y2,'r-', lw = 2, alpha = 0.8, color = "green"); 
plt.plot(X3,Y3,'r-', lw = 2, alpha = 0.8, color = "blue"); 
plt.plot(X4,Y4,'r-', lw = 2, alpha = 0.8, color = "magenta"); 
plt.plot(X5,Y5,'r-', lw = 2, alpha = 0.8, color = "cyan");

plt.legend( ('OT BER: 1*e-5', 'OT BER: 1*e-7', 'MHDR BER : 1*e-7','Double_Delay BER : 1*e-7',
             'Half_Delay BER : 1*e-7' ), loc = 'upper left', prop={'size': 15})
plt.xlabel("Simulation Time in Seconds", fontsize = 14, fontweight='bold')
plt.ylabel("Throughput in bits/sec", fontsize = 14, fontweight='bold')
plt.title("THROUGHPUT COMPARISON", color = "blue", fontsize = 20, fontweight='bold')
plt.show()
```
